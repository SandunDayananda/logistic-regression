```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    """Sigmoid activation function"""
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, theta):
    """Compute the cost for logistic regression"""
    m = len(y)
    h = sigmoid(X.dot(theta))
    cost = (1 / m) * (-y.T.dot(np.log(h)) - (1 - y).T.dot(np.log(1 - h)))
    return cost

def gradient_descent(X, y, theta, alpha, num_iters):
    """Perform gradient descent to learn theta"""
    m = len(y)
    costs = []

    for i in range(num_iters):
        h = sigmoid(X.dot(theta))
        error = h - y
        grad = (1 / m) * X.T.dot(error)
        theta -= alpha * grad
        cost = compute_cost(X, y, theta)
        costs.append(cost)

    return theta, costs

# Generate some random data
np.random.seed(0)
num_samples = 100
X = np.random.randn(num_samples, 2)
ones = np.ones((num_samples, 1))
X = np.hstack((ones, X))
true_theta = np.array([0.5, 0.5, 0.5])  # True parameters
y = np.random.randint(2, size=num_samples)

# Initialize theta and set hyperparameters
theta = np.zeros(X.shape[1])
alpha = 0.01
num_iters = 1000

# Run gradient descent
theta_final, costs = gradient_descent(X, y, theta, alpha, num_iters)

# Plot the cost against the number of iterations
plt.plot(range(num_iters), costs)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost')
plt.title('Cost vs Number of Iterations')
plt.show()
```
